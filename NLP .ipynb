```python
#printing first fourty words with highest occurence 

items = rider.word_counts.items()
items
```

```python
item = [item for item in items if item[0] not in stop_word]
```

```python
#sorting the items
from operator import itemgetter
sorted_item = sorted(item, key = itemgetter(1), reverse = True)
sorted_item
```

```python
#printing the first fourty words
first_20 = sorted_item[2:22]
import pandas as pd
df = pd.DataFrame(first_20, columns = ['Word', 'Count'])
```
